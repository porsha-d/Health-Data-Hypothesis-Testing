```python
#Question: Is the average cholesterol level in the sample significantly different from a known population mean 200 mg/dL?
```

```python
#H0:the average cholesterol level in the sample is not significantly different from a known population mean 200
#Ha:the average cholesterol level in the sample is significantly different from a known population mean 200
#one sample T-test
```

```python
from scipy.stats import ttest_1samp, t
```

```python
sample_mean=np.mean(merged_data['Total_Cholesterol'])
```

```python
print(sample_mean)
```

```python
sample_std=np.std(merged_data['Total_Cholesterol'], ddof=1)
```

```python
n=len(merged_data['Total_Cholesterol'])
```

```python
# Set confidence level
confidence_level = 0.95
alpha = 1 - confidence_level
```

```python
# Compute the critical t-value
t_critical = t.ppf(1 - alpha/2, df=n-1)
```

```python
#margin of error
margin_of_error = t_critical * (sample_std / np.sqrt(n))
```

```python
# Confidence interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error
```

```python
print(ci_lower, ci_upper)
```

```python
stat, p_value=ttest_1samp(merged_data['Total_Cholesterol'], 200)
```

```python
print(stat,p_value)
```

```python
#since p-value is less than 0.05 and confidence interval does not contain 200, we are confident that the average cholesterol level in the sample is significantly different from a known population mean 200 mg/dL. 
```

```python
from scipy.stats import ttest_ind, t
```

```python
sample_mean_male=np.mean(insulin_male)
```

```python
sample_mean_female=np.mean(insulin_female)
```

```python
sample_stdev_male=np.std((insulin_male), ddof=1)
```

```python
sample_stdev_female=np.std((insulin_female), ddof=1)
```

```python
n_male=len(insulin_male)
n_female=len(insulin_male)
print(n_male, n_female)
```

```python
# Standard error of the difference
se_diff = np.sqrt((sample_stdev_male**2 / n_male) + (sample_stdev_female**2 / n_female))
```

```python
#degrees of freedom
df=n_male+n_female-2
```

```python
#set confidence level
confidence=0.95
alpha=1-confidence
```

```python
#critical t-value
t_critical=t.ppf(1-alpha/2, df)
```

```python
#margin of error
margin_of_error=t_critical*se_diff
```

```python
mean_diff=sample_mean_male-sample_mean_female
print(mean_diff)
```

```python
#confidence interval
ci_lower=mean_diff-margin_of_error
ci_upper=mean_diff+margin_of_error
```

```python
print(ci_lower, ci_upper)
```

```python
print(stat, p_value)
```
